Solar irradiance can be direct (from the sun's beam) or diffuse (from scattering of the sun's beam). This scattering effect is caused by atmospheric conditions such as the presence of

#### Task 4.2: Seasons on Neptune ####

The planet Neptune has a similar tilt to that of Earth,
so has a similar seasonal pattern (but with seasons of different durations,
of course). It takes about 16 earth hours for Neptune to rotate around
its axis, and about 165 earth years to revolve around the sun.  Derive an equation for the number of earth hours of "daytime" on Neptune on each day of a Neptune year. You may assume that the Neptune year starts
on any Neptune day that you choose. You can choose an approximate latitude on Neptune for your model, or you can use latitude as a variable in your model. 

You may find it useful to examine the daytimes model discussed in Workshop 6 (for Earth).


```python
from math import sin, pi

hours_in_day = 16.0
days_in_year = 165

def daytime_hours(day_of_year):
    # a = hours_in_day / 2.0
    a = 5
    return clamp(a * sin(2*pi / days_in_year * (day_of_year)) + hours_in_day / 2.0)

for day in range(0, days_in_year):
    print(f"Hours at {day}: {daytime_hours(day)}")
```

Many organisms have different population growth rates in different seasons,

Consider a population of fish in which the change in biomass per month, $r_i$, in Month 
$i$ is given by


$$r_i=0.04+0.06\sin{\left(\frac{2\pi (i-9)}{12}\right)}$$


where $i$ is the month number between 0 and 11.
   1. Find the largest and smallest monthly change in biomass during the year, and 
the month numbers in which they occur. Interpret your answer.


   2. By hand, calculate the values of $r_0$ and $r_1$.


   3. In the following Python cell, write a program that uses a loop to calculate and print the values of $r_i$ for $i=0,1,\ldots, 11$. Check your answers for the first two months against your answers to Part 2.


   4. Explain what will happen to the values of $r_i$ for a month with number larger than 11.


   5. In the Python cell below, write a program that does all of the following:
        * Prints a suitable introductory message.
        * Prompts the user to enter the initial biomass in kg and the number of months.
        * Applies the growth rate $r_i$ for month $i$, and does this for the number of months specified
        * Prints the growth rate, and biomass each month
        * Prints the final biomass, with a suitable message.
        * Includes appropriate comments.


   6. Use your program to find the biomass of a population of fish after 24 months, given that the biomass is originally 500 kg. 

```python
# Write your code here
from math import sin, pi

print("Suitable introductory message")

initial = float(input("Initial biomass in kg: "))
num_months = int(input("How many months: "))

def r(i):
    return 0.04 + 0.06*sin(2*pi/12 * (i - 9))

final = initial
for i in range(0, num_months):
    final = final + r(i)

print(f"After {num_months} months starting at {initial}kg, the final biomass is: {final}kg")
```
